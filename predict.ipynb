```python
import numpy as np
```

```python
np.linspace(100,1200,12)
```

```python
n_estimators=[int(x) for x in np.linspace(100,1200,6)]
```

```python
print(n_estimators)
```

```python
max_features=["auto","sqrt"]
```

```python
max_depth=[x for x in np.linspace(5,30,12)]
```

```python
min_samples_split=[2,5,10,15,100]
```

```python
min_samples_leaf=[1,2,5,10]
```

```python
#gives best parameters to select the model for RandomForestRegressor method
from sklearn.model_selection import RandomizedSearchCV
```

```python
# we need to create the Grid 
random_grid={"n_estimators":n_estimators,
             "max_features":max_features,
             "max_depth":max_depth,
             "min_samples_split":min_samples_split,
             "min_samples_leaf":min_samples_leaf
}
```

```python
print(random_grid)
```

```python
rf=RandomForestRegressor()
```

```python
rf_random=RandomizedSearchCV(estimator=rf,param_distributions=random_grid,scoring="neg_mean_squared_error",n_iter=10,cv=5,verbose=2,random_state=11,n_jobs=1)
```

```python
rf_random.fit(x_train,y_train)
```

```python
predict=rf_random.predict(x_test)
```
